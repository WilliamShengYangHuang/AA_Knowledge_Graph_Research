```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Add some nodes to the graph
people = ["Alice", "Bob", "Charlie", "Dave", "Eve", "Frank", "Grace", "Heather", "Ivy", "Jasmine"]
G.add_nodes_from(people)

# Add some edges to the graph
G.add_edge("Alice", "Bob")
G.add_edge("Alice", "Dave")
G.add_edge("Bob", "Charlie")
G.add_edge("Charlie", "Dave")
G.add_edge("Dave", "Eve")
G.add_edge("Eve", "Frank")
G.add_edge("Frank", "Grace")
G.add_edge("Frank", "Heather")
G.add_edge("Grace", "Heather")
G.add_edge("Heather", "Ivy")
G.add_edge("Ivy", "Jasmine")
G.add_edge("Jasmine", "Alice")

# Visualize the graph
nx.draw(G, with_labels=True)
plt.show()
```

```python

import random
import string
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Add some nodes to the graph
people = []
for i in range(32):
    name = "".join(random.choices(string.ascii_uppercase, k=5))
    people.append(name)
    G.add_node(name)

# Add some edges to the graph
for i in range(len(people)):
    for j in range(i+1, len(people)):
        if set(people[i]) & set(people[j]):
            G.add_edge(people[i], people[j])

# Visualize the graph
plt.figure(figsize=(5, 5), dpi=150)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_size=4, node_size=10, node_color="orange", width=0.1, edge_color="grey")
plt.show()

```
